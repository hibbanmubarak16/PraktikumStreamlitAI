```python
%%writefile app.py
import streamlit as st
import pandas as pd
import plotly.express as px

# 1. Konfigurasi Halaman
st.set_page_config(page_title="Static Dashboard", layout="wide")

try:
    # 2. Load Data
    df = pd.read_csv('restaurants_tallinn.csv')

    st.title("Restaurant Analysis Dashboard (Static)")
    st.markdown("Dashboard ini menampilkan data tetap tanpa simulasi real-time.")

    # 3. Sidebar Filter (Interaktif tapi tidak otomatis update)
    st.sidebar.header("Filter Options")
    selected_cuisine = st.sidebar.multiselect(
        "Select Cuisine:",
        options=df["Cuisine"].unique(),
        default=df["Cuisine"].unique()[:3]
    )

    # Filter Data berdasarkan input
    df_filtered = df[df["Cuisine"].isin(selected_cuisine)]

    # 4. Metrics (KPI)
    kpi1, kpi2, kpi3 = st.columns(3)
    kpi1.metric("Total Restaurants", len(df_filtered))
    kpi2.metric("Average Score", round(df_filtered["Weighted_score"].mean(), 2))
    kpi3.metric("Expensive Places", len(df_filtered[df_filtered["Price_category"] == "expensive"]))

    st.divider()

    # 5. Charts
    col1, col2 = st.columns(2)

    with col1:
        st.subheader("Score Distribution by Cuisine")
        fig1 = px.box(df_filtered, x="Cuisine", y="Weighted_score", color="Price_category")
        st.plotly_chart(fig1, use_container_width=True)

    with col2:
        st.subheader("Price Category Count")
        fig2 = px.pie(df_filtered, names="Price_category", hole=0.4)
        st.plotly_chart(fig2, use_container_width=True)

    # 6. Data Table
    st.subheader("Detailed Data View")
    st.dataframe(df_filtered)

except FileNotFoundError:
    st.error("File 'restaurants_tallinn.csv' tidak ditemukan!")
```
